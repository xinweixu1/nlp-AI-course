### Step 1 - Preprocessing the news data 

````python
import pandas as pd
csv_path = '/Users/xinweixu/Dropbox/learn/Comp_Prog/nlp/data/project01-chinese-news/sqlResult_full.csv'
````

````python
# read in the content, skipping the lines can't be read

content = pd.read_csv(csv_path, delimiter=';', quotechar='"', escapechar='\\', error_bad_lines=False)
````

````python
content = content.fillna('')
````

````python
news_content = content['content'].tolist()
````

````python
news_content[:10]
````

````python
# define functions to clean the strings and cut to words

import jieba
import re

def cut(string): return ' '.join(jieba.cut(string))
def token(string):
    return re.findall(r'[\d|\w]+', string)
````

````python
token('这是一个测试\n\n\n')
````

````python
# filter out all redandunt strings in each article
news_content = [token(n) for n in news_content]
````

````python
news_content[:5]
````

````python
# join the sentences together in each article

news_content = [' '.join(n) for n in news_content]
````

````python
news_content[:5]
````

````python
news_content = [cut(n) for n in news_content]
````

````python
news_content[:5]
````

````python
# store as the Linesentence format readable by gensim word2vec
with open('/Users/xinweixu/Dropbox/learn/Comp_Prog/nlp/data/project01-chinese-news/news-sentences-cut.txt', 'w') as f:
    for n in news_content:
        f.write(n + '\n')
````

### Step 2.1 - Get word vectors

````python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
````

````python
line_sentences_path = '/Users/xinweixu/Dropbox/learn/Comp_Prog/nlp/data/project01-chinese-news/news-sentences-cut.txt'
````

````python
news_model = Word2Vec( LineSentence(line_sentences_path), min_count=5, workers=8)
````

````python
# save the model 

news_model.save('/Users/xinweixu/Dropbox/learn/Comp_Prog/nlp/data/project01-chinese-news/news_model')


# to load a model, use the following:
# news_model = gensim.models.Word2Vec.load('news_model')

# Advanced users can load a model and continue training it with more sentences:
#model = gensim.models.Word2Vec.load('path/to/mymodel')
#model.train(more_sentences)
````

````python
# also save the word vectors for future query
# here we save as .bin binary files since .bin files take less space

vectors_path = '/Users/xinweixu/Dropbox/learn/Comp_Prog/nlp/data/project01-chinese-news/news_model_word_vec.bin'
word_vectors = news_model.wv
word_vectors.save_word2vec_format(vectors_path, binary=True)

# to load binary word vectors into a model from a local directory, use the following:
#from gensim.models import KeyedVectors
#model = KeyedVectors.load_word2vec_format('path/to/<file_name>.bin', binary=True)
````

````python
# check out the POS labels

pos = []
for word in result_ch["sentences"][0]['tokens']:
    pos.append('{} ({})'.format(word['word'], word['pos']))
    
" ".join(pos)
````

````python
# filter out non-verbs

saying_verbs = []

for word in result_ch["sentences"][0]['tokens']:
    if word['pos'] == 'VV':
        saying_verbs.append(word['word']) 
    
````
